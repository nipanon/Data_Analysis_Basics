```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bitcoin = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
bitcoin.head()
```

```python
bitcoin['Close'].plot(figsize=(9, 6))
plt.show()
```

```python
bitcoin.index
```

```python
bitcoin['2019-09']['Close'].plot()
```

```python
bitcoin['2017':'2019']['Close'].plot()
```

```python
#Regroupement des données selon une fréquence ('M' = par mois)
bitcoin.loc['2017':'2019']['Close'].resample('M').plot()
plt.show()
```

```python
bitcoin.loc['2019', 'Close'].resample('W').mean().plot()
plt.show()
```

```python
bitcoin.loc['2019', 'Close'].resample('2W').std().plot()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
bitcoin.loc['2020', 'Close'].plot()
bitcoin.loc['2020', 'Close'].resample('M').mean().plot(label='moyenne par mois', lw=3, ls=':', alpha=0.8)
bitcoin.loc['2020', 'Close'].resample('W').mean().plot(label='moyenne par mois', lw=2, ls='--', alpha=0.8)
plt.legend()
plt.show()
```

```python
m = bitcoin['Close'].resample('W').agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize=(12, 8))
m['mean']['2020'].plot(label='moyenne par semaine')
plt.fill_between(m.index, m['max'], m['min'], alpha=0.2, label='min-max par semaine')

plt.legend()
plt.show()
```

## Moving Average

```python
plt.figure(figsize=(12, 8))
bitcoin.loc['2020-09', 'Close'].plot()
bitcoin.loc['2020-09', 'Close'].rolling(window=7).mean().plot(label='moving average', lw=3, ls=':', alpha=0.8)
bitcoin.loc['2020-09', 'Close'].rolling(window=7, center=True).mean().plot(label='moving average', lw=3, ls=':', alpha=0.8)

plt.legend()
plt.show()
```

## Exponential Weighted Function

```python
plt.figure(figsize=(12, 8))
bitcoin.loc['2020-09', 'Close'].plot()
bitcoin.loc['2020-09', 'Close'].rolling(window=7).mean().plot(label='moving average', lw=3, ls=':', alpha=0.8)
bitcoin.loc['2020-09', 'Close'].rolling(window=7, center=True).mean().plot(label='moving average', lw=3, ls=':', alpha=0.8)
bitcoin.loc['2020-09', 'Close'].ewm(alpha=0.6).mean().plot(label='moving average', lw=3, ls='--', alpha=0.8)

plt.legend()
plt.show()
```
